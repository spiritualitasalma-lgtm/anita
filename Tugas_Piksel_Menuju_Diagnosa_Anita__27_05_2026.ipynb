```python
# LAPORAN STUDI KASUS BIOMEDIS
# Judul: Analisis Morfologi Sel Darah Menggunakan Ekstraksi Ciri Bentuk (External Descriptors), untuk Otomatisasi Deteksi Acanthocyte Berbasis Sinyal Deret Waktu*

# 1.Latar Belakang
# Darah merupakan komponen vital yang membawa informasi penting mengenai kondisi kesehatan pasien.
# Pada kondisi normal, sel darah merah (*Erythrocyte*) berbentuk cakram bikonkaf halus simetris yang efisien dalam mengikat oksigen.
# Namun, beberapa penyakit sistemik atau kelainan genetik dapat merusak integritas membran luar sel.
# Salah satu bentuk deformasi yang krusial dipantau adalah **Acanthocyte** atau **Echinocyte,
# (sel darah merah berduri). Perubahan fisik berupa tonjolan taji (*crenation*) ini sering menjadi indikator klinis penting bagi penyakit,
# degeneratif saraf (seperti *Neuroacanthocytosis*), gangguan fungsi hati yang parah, malnutrisi, hingga gagal ginjal.
# Secara konvensional, penentuan jenis sel ini dilakukan oleh spesialis patologi menggunakan mikroskop manual.
# Metode ini memiliki kelemahan karena bersifat subjektif, memakan waktu lama, dan rentan terhadap kelelahan visual (human error). Oleh karena itu, diperlukan pendekatan otomatisasi berbasis *Computer Vision* biomedis yang mampu mengukur kerusakan morfologi sel secara objektif dan kuantitatif.

# 2. Pengertian
    # a: External Descriptors (Deskriptor Bentuk Luar):
    #  Metode dalam pemrosesan citra digital yang fokus pada ekstraksi karakteristik geometri berdasarkan koordinat batas/tepi objek (*contour*), bukan piksel bagian dalam objek.
    # b: Transformasi Polar ke Deret Waktu (*Time-Series*):
    # Proses matematika untuk membentangkan kontur melingkar sel menjadi sebuah sinyal linear satu dimensi berdasarkan sumbu sudut ($0^\circ$ hingga $360^\circ$) dan sumbu jarak (radius piksel ke pusat massa).
    # c: Standar Deviasi ($\sigma$) Kontur:
    # Metrik statistik yang digunakan sebagai parameter kuantitatif untuk mengukur seberapa jauh sebaran piksel tepi sel melenceng dari rata-rata bentuk ideal sel sehat.

# 3. Tujuan Studi Kasus
    # a: Segmentasi Otomatis: Memisahkan objek sel tunggal dari latar belakang citra mikroskopis yang penuh *noise* tekstur secara bersih.
    # b: Kuantifikasi Morfologi: Mengonversi karakteristik fisik sel darah yang semula dinilai secara kualitatif,
    # ("terlihat berduri") menjadi data numerik terukur (piksel deviasi).
    # C: Otomatisasi Diagnosa:Membangun sistem deteksi dini berbasis aturan, (*rule-based*) untuk menentukan status kesehatan sel darah (Normal vs Abnormal) tanpa intervensi manual.

# 4. Proses Implementasi & Langkah-Langkah.
# Proses transformasi dari data piksel mentah menuju keputusan klinis dilakukan melalui alur sebagai berikut:
 # a: [Citra Mikroskopis] ➔ [Gaussian Blur (Reduksi Noise)] ➔ [Otsu's Thresholding (Biner)]
 # b: [Hasil Diagnosa Otomatis] ⬅ [Analisis Standar Deviasi (σ)] ⬅ [Ekstraksi Kontur & Transformasi Polar]
    # 1. **Pra-proses Citra:** Mengurangi *noise* bintik-bintik putih acak (*Gaussian noise*) pada gambar menggunakan filter Gaussian halus agar tepi objek sel menjadi lebih kontinu saat diekstraksi.
    # 2. **Segmentasi Biner:** Menggunakan teknik *Otsu Thresholding* untuk menentukan ambang batas optimal, memisahkan piksel sel (putih) dengan latar belakang (hitam).
    # 3. **Ekstraksi Kontur:** Melacak koordinat tepi terluar objek sel darah guna mengisolasi geometri luarnya.
    # 4. **Pemetaan Deret Waktu:** Menghitung jarak geometri dari pusat massa sel ke setiap piksel kontur, lalu mengurutkannya searah jarum jam dari sudut $0^\circ$ hingga $360^\circ$.

# 5. Hasil Analisis Kuantitatif & Grafis.
# Berdasarkan visualisasi dan kalkulasi sistem komputer terhadap gambar simulasi sel berduri tersebut, diperoleh hasil kuantitatif sebagai berikut:
# Total Koordinat Piksel Diperiksa:** $592$ Titik Kontur.
# Rata-rata Radius Geometri Sel ($\mu$):** $90.14$ Piksel.
# Kuantifikasi Standar Deviasi ($\sigma$):** $8.32$ Piksel.

# Visualisasi Output Sistem.
# Karakteristik Sinyal:** Ketika dibentangkan secara linear, kontur sel menghasilkan grafik sinyal berwarna biru yang membentuk pola **11 puncak gelombang sinus tajam** yang mewakili 11 taji/duri pada membran sel.
# Penyimpangan Geometri:** Garis putus-putus putih (lingkaran ideal rata-rata) gagal berhimpit dengan kontur asli sel. Tingginya sebaran piksel yang menjauhi garis rata-rata ini tercermin dari area Standar Deviasi ($\sigma = 8.32$ px) yang melebihi ambang batas toleransi sel normal ($> 4.0$ px).

# 6. Kegunaan (Manfaat Klinis).
# Akurasi Tinggi & Objektif:** Menghilangkan bias penilaian subjektif antar laboratorium patologi karena keputusan diambil murni berdasarkan deviasi angka matematis.
# Skalabilitas Deteksi:** Algoritma ini dapat diterapkan pada sistem *High-Throughput Screening* untuk memeriksa ribuan sampel sel darah dalam hitungan detik.
# Pengarsipan Rekam Medis Digital:** Nilai deret waktu dan standar deviasi sel pasien dapat disimpan sebagai data riwayat klinis numerik yang efisien untuk memantau perkembangan efektivitas pengobatan dari waktu ke waktu.

# 7. Kesimpulan.
# Melalui parameter *external descriptors*, teknologi *Computer Vision* terbukti tidak lagi sekadar menampilkan gambar visual, melainkan mampu melakukan bedah kuantitatif terhadap struktur fisik objek biomedis. Nilai Standar Deviasi ($\sigma$) yang tinggi ($8.32$ piksel) menjadi basis pembuat keputusan (*decision-maker*) bagi komputer untuk menetapkan status **ABNORMAL (Acanthocyte/Echinocyte)**. Metode ini berhasil mengubah bentuk fisik sel darah yang kompleks menjadi representasi data deret waktu linear yang akurat dan mudah diinterpretasikan.

# 8. Penutup
# Integrasi antara pengolahan citra digital (*OpenCV* & *Skimage*), dengan analisis statistik terapan menandai lompatan besar dalam dunia teknik biomedis modern.
# Pendekatan *"Dari Piksel Menuju Diagnosa"* ini membuka peluang besar untuk pengembangan laboratorium cerdas masa depan. Dengan memanfaatkan deteksi dini otomatis yang handal, beban kerja tenaga medis dapat dikurangi secara signifikan, dan yang paling utama, keputusan klinis kritis bagi keselamatan pasien dapat diambil secara jauh lebih cepat dan akurat.
```

```python
# 1. Persiapan Gambar Simulasi Sel
# Karena kita bekerja di Google Colab, kode di bawah ini sudah dilengkapi fungsi otomatis,
# untuk membuat simulasi sel darah berduri (mirip dengan objek pada gambar acuan Anda) menggunakan formula matematika. Jadi, Anda bisa langsung mencobanya tanpa perlu mengunggah gambar luar terlebih dahulu.

# Coodingannya sbb:
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import regionprops, label

# 1. Membuat Citra Simulasi Sel Darah Berduri (Biomedis)
width, height = 400, 400
img_simulated = np.zeros((width, height), dtype=np.uint8)
center_x, center_y = width // 2, height // 2
base_radius = 90

# Membuat bentuk lingkaran dengan intervensi gelombang sinus untuk efek "duri/lekukan"
for y in range(height):
    for x in range(width):
        dx = x - center_x
        dy = y - center_y
        distance = np.sqrt(dx**2 + dy**2)

        if distance > 0:
            angle = np.arctan2(dy, dx)
            # Menambahkan 11 duri di sekeliling sel
            crenation = 12 * np.sin(11 * angle)
            if distance <= (base_radius + crenation):
                img_simulated[y, x] = 180  # Warna abu-abu sel

# Menambahkan noise Gaussian halus agar teksturnya mirip hasil mikroskop asli
noise = np.random.normal(0, 15, img_simulated.shape).astype(np.uint8)
img_cells = cv2.add(img_simulated, noise)

# Tampilkan gambar awal
plt.figure(figsize=(5, 5))
plt.imshow(img_cells, cmap='gray')
plt.title("Citra Mikroskopis Sel (Simulasi)")
plt.axis('off')
plt.show()
```

```python
# 2. Alur Pemrosesan Citra (Image Processing Pipeline)
# Di bawah ini adalah langkah-langkah preprocessing (segmentasi biner dan pembersihan morfologi)
# untuk memisahkan objek sel dari latar belakangnya secara sempurna.

# Coodingannya sbb:
# 2. Filtering untuk mereduksi noise tekstur mikroskop
img_blur = cv2.GaussianBlur(img_cells, (9, 9), 0)

# 3. Segmentasi dengan Otsu Thresholding
_, img_thresh = cv2.threshold(img_blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 4. Operasi Morfologi (Closing) untuk menghaluskan area dalam sel
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
img_morph = cv2.morphologyEx(img_thresh, cv2.MORPH_CLOSE, kernel)

# Visualisasi Tahapan Proses
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(img_blur, cmap='gray')
axes[0].set_title("1. Gaussian Blur")
axes[0].axis('off')

axes[1].imshow(img_thresh, cmap='gray')
axes[1].set_title("2. Otsu Thresholding")
axes[1].axis('off')

axes[2].imshow(img_morph, cmap='gray')
axes[2].set_title("3. Morphological Cleaned")
axes[2].axis('off')
plt.show()
```

```python
# 3. Ekstraksi Kontur dan Perhitungan External Descriptors
# Langkah ini mengekstrak deskriptor batas luar sel menggunakan perpaduan fungsi cv2.findContours dari OpenCV,
# dan properti tingkat lanjut milik skimage.measure.regionprops.
# Untuk menambahkan fitur Visualisasi Hasil Deteksi langsung pada kode ekstraksi kontur yang Anda berikan, kita perlu menambahkan pustaka Matplotlib (plt) dan beberapa fungsi bawaan OpenCV (cv2.rectangle dan cv2.drawContours).
# Garis kontur luarnya akan kita warnai Magenta, dan kotak batas tegaknya (Bounding Box) akan kita warnai Biru, persis seperti tampilan standar pada modul biomedis "Dari Piksel Menuju Diagnosa".

# Berikut adalah kode lengkap yang sudah digabungkan dengan fungsi visualisasi gambar ini Coodingannya sbb:
# import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import regionprops, label

# 1. PENDETAAN CITRA BINAR MANDIRI (img_morph)
# Membuat background hitam berukuran 400x400 piksel (Citra 1-Kanal / Gray)
width, height = 400, 400
img_morph = np.zeros((width, height), dtype=np.uint8)
center_x, center_y = width // 2, height // 2
base_radius = 90

# Menggambar sel berduri buatan pada matriks citra biner
for y in range(height):
    for x in range(width):
        dx, dy = x - center_x, y - center_y
        distance = np.sqrt(dx**2 + dy**2)
        if distance > 0:
            angle = np.arctan2(dy, dx)
            crenation = 12 * np.sin(11 * angle) # Modul 11 taji sel
            if distance <= (base_radius + crenation):
                img_morph[y, x] = 255 # Objek diberi warna putih


# 2. EKSTRAKSI KONTUR DAN PARAMETER GEOMETRI
contours, _ = cv2.findContours(img_morph, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
main_contour = max(contours, key=cv2.contourArea)

# Ambil titik koordinat kotak pembungkus objek (Bounding Box)
x, y, w, h = cv2.boundingRect(main_contour)

# 3. IMPLEMENTASI FITUR GAMBAR.
# LANGKAH A: Ubah citra biner (1 Kanal) ke format 3 Kanal (BGR)
# Tanpa perintah ini, goresan kuas digital hanya akan menjadi warna abu-abu/putih datar.
img_visual = cv2.cvtColor(img_morph, cv2.COLOR_GRAY2BGR)

# LANGKAH B: Menggunakan cv2.rectangle sebagai kuas digital pembentuk kotak batas tegak
# Kita goreskan warna BIRU pada koordinat penampung. Format BGR -> (255, 0, 0), tebal garis = 2 px.
cv2.rectangle(img_visual, (x, y), (x + w, y + h), (255, 0, 0), 2)

# LANGKAH C: Menggunakan cv2.drawContours sebagai kuas penjiplak setiap piksel tepi kontur
# Kita goreskan warna MAGENTA sepanjang untaian koordinat sel. Format BGR -> (255, 0, 255), tebal = 2 px.
cv2.drawContours(img_visual, [main_contour], -1, (255, 0, 255), 2)

# 4. MENAMPILKAN HASIL GAMBAR (OUTPUT MATPLOTLIB)
plt.figure(figsize=(6, 6))
# Konversi BGR OpenCV ke RGB Matplotlib agar warna Biru & Magenta tidak tertukar saat dirender
plt.imshow(cv2.cvtColor(img_visual, cv2.COLOR_BGR2RGB))

plt.title("TEMUAN HASIL GAMBAR\n(Implementasi cv2.cvtColor + cv2.rectangle + cv2.drawContours)",
          fontsize=12, fontweight='bold', pad=12)
plt.axis('off') # Menyembunyikan penggaris angka koordinat agar citra bersih
plt.show()
```

```python
# 4. Visualisasi Hasil Akhir Sesuai Gambar Referensi
# Kode ini akan menampilkan visualisasi akhir berupa garis kontur berwarna magenta (pink)
# melingkari tepi sel serta bounding box berwarna biru, persis seperti ilustrasi studi kasus "Dari Piksel Menuju Diagnosa".

# Coodingannya sbb:
# 7. Menggambar Visualisasi Kontur dan Bounding Box
img_result = cv2.cvtColor(img_morph, cv2.COLOR_GRAY2BGR)

# Gambar Bounding Box (Biru)
cv2.rectangle(img_result, (x, y), (x + w, y + h), (255, 0, 0), 2)
# Gambar Kontur Eksternal (Magenta / Pink)
cv2.drawContours(img_result, [main_contour], -1, (255, 0, 255), 2)

# Menampilkan Hasil Plot Akhir
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB))
plt.title("Hasil Ekstraksi Fitur Morfologi Sel")
plt.axis('off')
plt.show()

# 8. Menampilkan Hasil Print Out Data Kuantitatif Medis
print("+" + "="*47 + "+")
print("|EKSTRAKSI CIRI GEOMETRI SEL |")
print("+" + "="*47 + "+")
print(f"| Luas Objek (Area)       : {area:<10.1f} piksel|")
print(f"| Keliling (Perimeter)    : {perimeter:<10.1f} piksel|")
print(f"| Aspect Ratio            : {aspect_ratio:<10.4f}|")
print(f"| Circularity (Roundness) : {circularity:<10.4f}|")
print(f"| Solidity (Kepadatan)    : {solidity:<10.4f}|")
print(f"| Eccentricity (Kelonjongan): {eccentricity:<10.4f}|")
print("+" + "="*47 + "+")
```

```python
# STEP 6: OTOMATISASI DIAGNOSA MEDIS
# Inisialisasi variabel string untuk menyimpan teks analisis
# Untuk menampilkan hasil Interpretasi Diagnosa Medis secara maksimal layaknya sebuah dasbor atau laporan laboratorium klinis yang profesional, kita bisa memanfaatkan fungsi visualisasi teks kustom langsung di dalam kanvas Matplotlib.

# Coodingan sbb:
import cv2  # Perbaikan: Menambahkan import cv2 yang sempat terlewat
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import regionprops, label

# 1. PREPARASI DATA SIMULASI (Asumsi Nilai Parameter Pengukuran)
width, height = 400, 400
img_morph = np.zeros((width, height), dtype=np.uint8)
center_x, center_y = width // 2, height // 2
base_radius = 90

# Simulasi pembuatan sel darah merah berduri (Acanthocyte)
for y in range(height):
    for x in range(width):
        dx, dy = x - center_x, y - center_y
        distance = np.sqrt(dx**2 + dy**2)
        if distance > 0:
            angle = np.arctan2(dy, dx)
            crenation = 12 * np.sin(11 * angle)
            if distance <= (base_radius + crenation):
                img_morph[y, x] = 255

# Deteksi Kontur dan Bounding Box
contours, _ = cv2.findContours(img_morph, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
main_contour = max(contours, key=cv2.contourArea)
x, y, w, h = cv2.boundingRect(main_contour)

# Hitung nilai descriptor geometri
area = cv2.contourArea(main_contour)
perimeter = cv2.arcLength(main_contour, True)
aspect_ratio = float(w) / h
circularity = (4 * np.pi * area) / (perimeter ** 2)

labeled_image = label(img_morph)
regions = regionprops(labeled_image)
main_region = max(regions, key=lambda r: r.area)
solidity = main_region.solidity
eccentricity = main_region.eccentricity

# 2. LOGIKA OTOMATISASI DIAGNOSA MEDIS (Sistem Pakar Rule-Based)
analisis_solidity = ""
analisis_circularity = ""
kesimpulan_sistem = ""
status_sel = ""

if solidity >= 0.96:
    analisis_solidity = "Normal (Tepian sel rata, tidak ditemukan struktur duri)."
else:
    analisis_solidity = f"Abnormal/Rendah ({solidity:.4f}). Indikasi lekukan tajam pada membran."

if circularity >= 0.85:
    analisis_circularity = "Normal (Bentuk sel melingkar secara simetris)."
else:
    analisis_circularity = f"Abnormal/Rendah ({circularity:.4f}). Keliling sel membengkak akibat duri."

if eccentricity < 0.6:
    if solidity < 0.96 and circularity < 0.85:
        status_sel = "ABNORMAL (Acanthocyte / Echinocyte)"
        kesimpulan_sistem = (
            "Kombinasi nilai Solidity dan Circularity yang rendah,\ndidukung oleh Eccentricity "
            "yang rendah (tetap bulat,\nbukan elips), mengonfirmasi secara otomatis bahwa sel\n"
            "darah merah ini mengalami crenation dan tergolong\nsebagai Sel Berduri (Acanthocyte/Echinocyte)."
        )
    else:
        status_sel = "NORMAL (Discocyte)"
        kesimpulan_sistem = "Sel berbentuk cakram bulat normal yang simetris\ntanpa kelainan morfologi luar."
else:
    if circularity < 0.6:
        status_sel = "ABNORMAL (Sickle Cell / Drepanocyte)"
        kesimpulan_sistem = "Nilai Eccentricity yang tinggi menunjukkan pemanjangan\nekstrem, mengarah pada indikasi Anemia Sel Sabit."
    else:
        status_sel = "ABNORMAL (Ovalocyte)"
        kesimpulan_sistem = "Sel berbentuk elips/oval secara konsisten,\nmengarah pada kelainan Ovalositosis."

# 3. VISUALISASI MAKSIMAL: KOMBINASI GAMBAR & LAPORAN DIGITAL
# Menyiapkan gambar anotasi kontur (Magenta) dan Bounding Box (Biru)
img_visual = cv2.cvtColor(img_morph, cv2.COLOR_GRAY2BGR)
cv2.rectangle(img_visual, (x, y), (x + w, y + h), (255, 0, 0), 2)
cv2.drawContours(img_visual, [main_contour], -1, (255, 0, 255), 2)

# Membuat subplots lebar (16x7 inch) untuk mengakomodasi gambar & teks rekam medis
fig, axes = plt.subplots(1, 2, figsize=(16, 7), gridspec_kw={'width_ratios': [1, 1.2]})
fig.patch.set_facecolor('#f4f6f9') # Mengubah background luar agar bernuansa clean-lab

# --- SUBPLOT 1: TAMPILAN FISIK OBJEK ---
axes[0].imshow(cv2.cvtColor(img_visual, cv2.COLOR_BGR2RGB))
axes[0].set_title("CITRA HASIL EKSTRAKSI CIRI BENTUK", fontsize=12, fontweight='bold', color='#1a202c', pad=15)
axes[0].axis('off')

# --- SUBPLOT 2: LEMBAR DIAGNOSA MEDIS DIGITAL ---
axes[1].axis('off')
# Set warna background area teks laporan berdasarkan status kesehatan sel
bg_color = '#fff5f5' if "ABNORMAL" in status_sel else '#f0fff4'
border_color = '#e53e3e' if "ABNORMAL" in status_sel else '#38a169'

# Membuat kotak latar belakang untuk laporan teks
axes[1].text(-0.05, -0.05, "", bbox=dict(boxstyle='round,pad=2', facecolor=bg_color, edgecolor=border_color, lw=2))

# Menyusun teks laporan formal klinis
report_title = "MEDICAL DIAGNOSIS REPORT (COMPUTER VISION)"
status_text = f"STATUS EVALUASI: {status_sel}"

# Menulis teks ke dalam kanvas gambar secara rapi dan presisi
axes[1].text(0.0, 0.95, report_title, fontsize=14, fontweight='bold', color='#2d3748')
axes[1].text(0.0, 0.90, "="*60, fontsize=11, color='#718096')

axes[1].text(0.0, 0.80, status_text, fontsize=14, fontweight='bold', color=border_color)
axes[1].text(0.0, 0.75, "-"*60, fontsize=11, color='#718096')

axes[1].text(0.0, 0.65, f"[•] Analisis Solidity        : {analisis_solidity}", fontsize=11, color='#2d3748')
axes[1].text(0.0, 0.55, f"[•] Analisis Circularity     : {analisis_circularity}", fontsize=11, color='#2d3748')
axes[1].text(0.0, 0.45, f"[•] Kelonjongan (Eccentricity): {eccentricity:.4f} ({'Kategori: Bulat' if eccentricity < 0.6 else 'Kategori: Elips'})", fontsize=11, color='#2d3748')

axes[1].text(0.0, 0.38, "-"*60, fontsize=11, color='#718096')
axes[1].text(0.0, 0.30, "[KESIMPULAN AKHIR SISTEM]:", fontsize=11, fontweight='bold', color='#2d3748')
axes[1].text(0.0, 0.08, kesimpulan_sistem, fontsize=11, color='#2d3748', linespacing=1.5)
axes[1].text(0.0, 0.00, "="*60, fontsize=11, color='#718096')

plt.tight_layout()
plt.show()

# Cetak log akhir di konsol terbawah sebagai konfirmasi sukses
print("\n>>> Pemrosesan Dasbor Laporan Terintegrasi Berhasil Ditampilkan <<<")
```

```python
# 1. Kode Lengkap: Pemrosesan, Deteksi Duri, & Analisis Kuantitatif
# import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import label, regionprops


# 1. GENERATE SIMULASI SEL BERDURI (Sama seperti tahap sebelumnya)
width, height = 400, 400
img_base = np.zeros((width, height), dtype=np.uint8)
center = (width // 2, height // 2)

for y in range(height):
    for x in range(width):
        dx, dy = x - center[0], y - center[1]
        distance = np.sqrt(dx**2 + dy**2)
        if distance > 0:
            angle = np.arctan2(dy, dx)
            # Membuat 11 duri/crenation
            crenation = 12 * np.sin(11 * angle)
            if distance <= (90 + crenation):
                img_base[y, x] = 255

# Pra-proses pembersihan tepi
img_blur = cv2.GaussianBlur(img_base, (5, 5), 0)
_, img_thresh = cv2.threshold(img_blur, 127, 255, cv2.THRESH_BINARY)


# 2. EKSTRAKSI KONTUR DAN CONVEX HULL (BENTUK IDEAL TANPA DURI)
contours, _ = cv2.findContours(img_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
main_contour = max(contours, key=cv2.contourArea)

# Hitung Convex Hull (Bentuk ideal penutup luar sel jika sel itu normal/rata)
hull = cv2.convexHull(main_contour, returnPoints=True)

# 3. MENGUKUR PENYIMPANGAN PIKSEL SECARA KUANTITATIF (QUANTITATIVE DEVIATION)
# Menyimpan data penyimpangan jarak jarak dari setiap piksel kontur ke pusat sel
distances_to_center = []
deviations = []

for point in main_contour:
    pt_x, pt_y = point[0][0], point[0][1]
    # Jarak piksel kontur ke pusat sel
    dist = np.sqrt((pt_x - center[0])**2 + (pt_y - center[1])**2)
    distances_to_center.append(dist)

    # Kuantifikasi Penyimpangan: Jarak absolut piksel kontur asli ke garis Convex Hull terdekat
    # Menggunakan fungsi cv2.pointPolygonTest
    dist_to_hull = abs(cv2.pointPolygonTest(hull, (float(pt_x), float(pt_y)), True))
    deviations.append(dist_to_hull)

# Cari statistik penyimpangan piksel
max_deviation = np.max(deviations)
mean_deviation = np.mean(deviations)

# 4. VISUALISASI HASIL GAMBAR & GRAFIK PENYIMPANGAN
img_rgb = cv2.cvtColor(img_thresh, cv2.COLOR_GRAY2BGR)

# Gambar bentuk ideal / Convex Hull (Garis Hijau)
cv2.drawContours(img_rgb, [hull], -1, (0, 255, 0), 2)
# Gambar bentuk asli berduri / Kontur (Garis Magenta)
cv2.drawContours(img_rgb, [main_contour], -1, (255, 0, 255), 2)

# Plot Hasil Akhir Berdampingan
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Subplot 1: Citra Hasil Pemisahan Otomatis
axes[0].imshow(img_rgb)
axes[0].set_title("Pemisahan Otomatis & Komparasi Bentuk\n(Magenta: Kontur Asli | Hijau: Ideal/Convex Hull)")
axes[0].axis('off')

# Subplot 2: Grafik Kuantifikasi Penyimpangan Piksel
axes[1].plot(deviations, color='crimson', linewidth=1.5)
axes[1].fill_between(range(len(deviations)), deviations, color='crimson', alpha=0.2)
axes[1].set_title("Grafik Fluktuasi Penyimpangan Piksel Terhadap Bentuk Ideal")
axes[1].set_xlabel("Indeks Titik Piksel Sepanjang Kontur")
axes[1].set_ylabel("Nilai Penyimpangan Jarak (Piksel)")
axes[1].grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()

# 5. PRINT OUT LAPORAN KUANTITATIF
print("\n" + "="*55)
print("     LAPORAN PENYIMPANGAN MORFOLOGI PIKSEL SECARA REAL     ")
print("="*55)
print(f"[*] Total Titik Piksel Kontur Diperiksa : {len(main_contour)} Titik")
print(f"[*] Rata-rata Penyimpangan Piksel       : {mean_deviation:.2f} piksel")
print(f"[*] Penyimpangan Piksel Maksimum (Duri) : {max_deviation:.2f} piksel")
print("-"*55)
if mean_deviation > 2.0:
    print("KESIMPULAN: Sel terdeteksi mengalami deformasi struktur.\n"
          "Fluktuasi grafik menunjukkan pola duri (crenation) yang berulang.")
else:
    print("KESIMPULAN: Sel normal. Tepian kontur sel berhimpit dengan garis ideal.")
print("="*55)
```

```python
# Nilai Tambah Medis
# 1. Kode Lengkap Konversi Kontur ke Data Deret Waktu (Time-Series)
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. MEMBUAT CITRA SIMULASI SEL DARAH BERDURI (ACANTHOCYTE)
width, height = 400, 400
img_base = np.zeros((width, height), dtype=np.uint8)
center_x, center_y = width // 2, height // 2
base_radius = 90

# Generasi bentuk sel dengan 11 duri menggunakan fungsi sinus
for y in range(height):
    for x in range(width):
        dx, dy = x - center_x, y - center_y
        distance = np.sqrt(dx**2 + dy**2)
        if distance > 0:
            angle = np.arctan2(dy, dx)
            crenation = 12 * np.sin(11 * angle)  # 11 buah duri
            if distance <= (base_radius + crenation):
                img_base[y, x] = 255

# Smoothing tepi citra
img_blur = cv2.GaussianBlur(img_base, (5, 5), 0)
_, img_thresh = cv2.threshold(img_blur, 127, 255, cv2.THRESH_BINARY)

# 2. EKSTRAKSI KONTUR TERLUAR
contours, _ = cv2.findContours(img_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
main_contour = max(contours, key=cv2.contourArea)

# 3. TRANSFORMASI KONTUR MENJADI DATA DERET WAKTU (TIME-SERIES)
# Mengubah koordinat Kartesian (X, Y) kontur menjadi koordinat Polar (Sudut, Jarak)
time_series_data = []
angles = []

for point in main_contour:
    pt_x, pt_y = point[0][0], point[0][1]

    # Hitung jarak (radius) dari pusat ke titik kontur
    r = np.sqrt((pt_x - center_x)**2 + (pt_y - center_y)**2)

    # Hitung sudut dalam derajat (-180 hingga 180 -> diubah ke 0 hingga 360)
    angle_rad = np.arctan2(pt_y - center_y, pt_x - center_x)
    angle_deg = np.degrees(angle_rad)
    if angle_deg < 0:
        angle_deg += 360

    time_series_data.append(r)
    angles.append(angle_deg)

# Urutkan data berdasarkan urutan sudut 0 - 360 derajat agar menjadi deret waktu yang rapi
sorted_indices = np.argsort(angles)
sorted_angles = np.array(angles)[sorted_indices]
sorted_distances = np.array(time_series_data)[sorted_indices]

# 4. STATISTIK MEDIS LANJUTAN: HITUNG STANDAR DEVIASI (σ)
mean_radius = np.mean(sorted_distances)
std_deviation = np.std(sorted_distances)  # Standar Deviasi Kuantitatif

# 5. VISUALISASI HASIL CITRA & GRAFIK DERET WAKTU MEDIS
img_result = cv2.cvtColor(img_thresh, cv2.COLOR_GRAY2BGR)
# Gambar kontur asli (Magenta)
cv2.drawContours(img_result, [main_contour], -1, (255, 0, 255), 3)
# Gambar lingkaran referensi rata-rata ideal (Garis Putih Putus-putus)
cv2.circle(img_result, (center_x, center_y), int(mean_radius), (255, 255, 255), 2, lineType=cv2.LINE_AA)

# Plotting Dashboard Utama
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Subplot 1: Deteksi Geometri Fisik Sel
axes[0].imshow(cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB))
axes[0].set_title("Analisis Fisik Sel Darah Berduri\n(Magenta: Tepi Sel | Putih: Jari-jari Rata-rata)", fontsize=12)
axes[0].axis('off')

# Subplot 2: Representasi Signal Deret Waktu (Time-Series Data)
axes[1].plot(sorted_angles, sorted_distances, color='blue', linewidth=2, label='Sinyal Kontur Sel')
axes[1].axhline(mean_radius, color='white', linestyle='--', linewidth=2, label=f'Rata-rata Radius ({mean_radius:.1f} px)')

# Efek visual area standar deviasi (rentang variasi duri)
axes[1].fill_between(sorted_angles, mean_radius - std_deviation, mean_radius + std_deviation,
                     color='yellow', alpha=0.2, label=f'Area Standar Deviasi (σ = {std_deviation:.2f})')

axes[1].set_title("Sinyal Deret Waktu (Time-Series) Kerusakan Membran Sel", fontsize=12)
axes[1].set_xlabel("Sudut Orientasi Melingkar Kontur (0° hingga 360°)", fontsize=10)
axes[1].set_ylabel("Jarak Piksel ke Pusat Sel (Radius)", fontsize=10)
axes[1].set_xlim(0, 360)
axes[1].set_ylim(60, 120)
axes[1].grid(True, linestyle=':', alpha=0.6)
axes[1].legend(loc='upper right')

plt.tight_layout()
plt.show()

# 6. PENUTUP & KESIMPULAN STATISTIK
print("\n" + "="*65)
print("       RINGKASAN STATISTIK BIOMEDIS (TIME-SERIES ANALYSIS)")
print("="*65)
print(f"[*] Rata-rata Jari-jari Sel (Mean Radius) : {mean_radius:.2f} piksel")
print(f"[*] Nilai Standar Deviasi (Std Deviation) : {std_deviation:.2f} piksel")
print("-"*65)
if std_deviation > 4.0:
    print(f"DIAGNOSA KLINIK: SEL ABNORMAL (Kategori: Echinocyte/Acanthocyte).\n"
          f"Nilai Standar Deviasi σ ({std_deviation:.2f}) yang tinggi menunjukkan fluktuasi "
          f"jarak yang ekstrem pada dinding sel, mengonfirmasi struktur berduri.")
else:
    print(f"DIAGNOSA KLINIK: SEL NORMAL (Kategori: Discocyte).\n"
          f"Nilai Standar Deviasi σ ({std_deviation:.2f}) mendekati 0, menandakan permukaan sel halus.")
print("="*65)
```

```python
# Kode Program Lengkap (Visualisasi & Output Closing Teks)
# import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. GENERASI CITRA SIMULASI SEL DARAH BERDURI (ACANTHOCYTE)
width, height = 400, 400
img_base = np.zeros((width, height), dtype=np.uint8)
center_x, center_y = width // 2, height // 2
base_radius = 90

# Membuat bentuk sel darah berduri (11 duri) menggunakan intervensi gelombang sinus
for y in range(height):
    for x in range(width):
        dx, dy = x - center_x, y - center_y
        distance = np.sqrt(dx**2 + dy**2)
        if distance > 0:
            angle = np.arctan2(dy, dx)
            crenation = 12 * np.sin(11 * angle)  # Modulasi 11 puncak duri
            if distance <= (base_radius + crenation):
                img_base[y, x] = 180  # Nilai keabuan matriks sel

# Menambahkan noise Gaussian halus dan smoothing tepi untuk efek mikroskop realistik
noise = np.random.normal(0, 10, img_base.shape).astype(np.uint8)
img_noisy = cv2.add(img_base, noise)
img_blur = cv2.GaussianBlur(img_noisy, (5, 5), 0)
_, img_thresh = cv2.threshold(img_blur, 100, 255, cv2.THRESH_BINARY)

# 2. EKSTRAKSI KONTUR TERLUAR (EXTERNAL DESCRIPTORS)
contours, _ = cv2.findContours(img_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
main_contour = max(contours, key=cv2.contourArea)

# 3. TRANSFORMASI MATEMATIS KONTUR MENJADI DATA DERET WAKTU (TIME-SERIES)
time_series_data = []
angles = []

for point in main_contour:
    pt_x, pt_y = point[0][0], point[0][1]

    # Kuantifikasi Jarak (Radius) dari pusat geometri ke piksel tepi kontur
    r = np.sqrt((pt_x - center_x)**2 + (pt_y - center_y)**2)

    # Hitung sudut arah (menghasilkan rentang 0 hingga 360 derajat)
    angle_rad = np.arctan2(pt_y - center_y, pt_x - center_x)
    angle_deg = np.degrees(angle_rad)
    if angle_deg < 0:
        angle_deg += 360

    time_series_data.append(r)
    angles.append(angle_deg)

# Mengurutkan deret data berdasarkan kronologi sudut 0° - 360°
sorted_indices = np.argsort(angles)
sorted_angles = np.array(angles)[sorted_indices]
sorted_distances = np.array(time_series_data)[sorted_indices]

# 4. KOMPUTASI STATISTIK LANJUTAN: VARIANS & STANDAR DEVIASI (σ)
mean_radius = np.mean(sorted_distances)
std_deviation = np.std(sorted_distances)  # Akar varians parameter penyimpangan

# 5. VISUALISASI EKSPEKTASI HASIL GAMBAR (2 PANEL BERDAMPINGAN)
img_result = cv2.cvtColor(img_thresh, cv2.COLOR_GRAY2BGR)

# Panel Kiri: Menggambar Kontur Asli (Magenta) & Lingkaran Rata-rata Rerata (Putih)
cv2.drawContours(img_result, [main_contour], -1, (255, 0, 255), 2)
cv2.circle(img_result, (center_x, center_y), int(mean_radius), (255, 255, 255), 2, lineType=cv2.LINE_AA)

# Inisialisasi Kanvas Gambar Berdampingan menggunakan Matplotlib
fig, axes = plt.subplots(1, 2, figsize=(16, 6.5))

# --- PANEL KIRI: CITRA HASIL DETEKSI FISIK ---
axes[0].imshow(cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB))
axes[0].set_title("PANEL KIRI: Citra Hasil Deteksi Fisik\n(Garis Magenta: Kontur Berduri | Putih Putus-putus: Rerata Radius Ideal)", fontsize=12, fontweight='bold', pad=10)
axes[0].axis('off')

# --- PANEL KANAN: GRAFIK SINYAL DERET WAKTU (TIME-SERIES) ---
axes[1].plot(sorted_angles, sorted_distances, color='blue', linewidth=2, label='Sinyal Kontur Sel (Radius)')
axes[1].axhline(mean_radius, color='black', linestyle='--', linewidth=2, label=f'Rata-rata Radius ({mean_radius:.1f} px)')

# Visualisasi area Standar Deviasi (Zona Transparan Kuning)
axes[1].fill_between(sorted_angles, mean_radius - std_deviation, mean_radius + std_deviation,
                     color='yellow', alpha=0.3, label=f'Zona Standar Deviasi (σ = {std_deviation:.2f} px)')

axes[1].set_title("PANEL KANAN: Grafik Sinyal Deret Waktu (Time-Series)\n(Membentangkan Geometri Lingkaran Kontur Menjadi Sinyal Linear)", fontsize=12, fontweight='bold', pad=10)
axes[1].set_xlabel("Sudut Orientasi Melingkar Kontur (0° sampai 360°)", fontsize=10)
axes[1].set_ylabel("Jarak Piksel Tepi ke Pusat Massa Sel (Radius)", fontsize=10)
axes[1].set_xlim(0, 360)
axes[1].set_ylim(60, 120)
axes[1].grid(True, linestyle=':', alpha=0.6)
axes[1].legend(loc='upper right')

plt.tight_layout()
plt.show()  # Perintah untuk menampilkan visualisasi di Google Colab

# 6. OUTPUT TEKS CLOSING (TERMINAL KONSOL REKAM MEDIS)
print("\n" + "="*75)
print("         HASIL EVALUASI BIOMEDIS AKHIR (DIAGNOSA BERBASIS KOMPUTASI)       ")
print("="*75)
print(f"[*] Total Koordinat Piksel Diperiksa   : {len(main_contour)} Titik Kontur")
print(f"[*] Rata-rata Radius Geometri Sel     : {mean_radius:.2f} Piksel")
print(f"[*] Kuantifikasi Standar Deviasi (σ)   : {std_deviation:.2f} Piksel")
print("-"*75)

# Rule-based thresholding otomatis untuk rekam medis
if std_deviation > 4.0:
    print("KESIMPULAN DIAGNOSA OTOMATIS:")
    print(">> [STATUS: ABNORMAL (Acanthocyte / Echinocyte)] <<")
    print(f"Justifikasi Medis: Nilai Standar Deviasi σ sebesar {std_deviation:.2f} piksel (> 4.0 px)\n"
          "menunjukkan fluktuasi dispersi jarak yang ekstrem dari pusat sel ke membran terluar.\n"
          "Sinyal deret waktu mengonfirmasi adanya deformasi struktural berupa taji/duri \n"
          "secara periodik di sepanjang dinding sel darah merah pasien.")
else:
    print("KESIMPULAN DIAGNOSA OTOMATIS:")
    print(">> [STATUS: NORMAL (Discocyte)] <<")
    print(f"Justifikasi Medis: Nilai Standar Deviasi σ sebesar {std_deviation:.2f} piksel (< 4.0 px)\n"
          "menandakan jarak antar-piksel terluar sel konsisten seragam. Membran luar sel\n"
          "bersifat halus, simetris, bebas dari anomali taji, dan dikategorikan sehat.")
print("="*75)
print(">>> Pemrosesan 'Dari Piksel Menuju Diagnosa' Selesai Secara Kuantitatif <<<")
```
