```python
def tampilkan(original, filtered, filter_name):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4), sharex=True,sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('original')
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(filter_name)
    ax2.axis('off')
```

```python
from skimage.morphology import erosion, dilation, opening, closing, white_tophat
from skimage.morphology import black_tophat, skeletonize, convex_hull_image
from skimage.morphology import disk

# Membuat Structuring Element (SE) berbentuk lingkaran dengan radius 5
selem = disk(3)

# Menerapkan operasi Erosi pada citra original menggunakan SE yang telah dibuat
eroded = erosion(orig_phantom, selem)

# Menampilkan perbandingan antara citra asli dan citra hasil erosi
tampilkan(orig_phantom, eroded, 'erosion')

# Memeriksa matriks dari Structuring Element
selem
```

```python
from skimage.morphology import erosion, dilation, opening, closing, white_tophat
from skimage.morphology import black_tophat, skeletonize, convex_hull_image
from skimage.morphology import disk

# Mempersiapkan elemen struktur (Structuring Element)
selem = disk(3)

# Melakukan operasi Erosi pada citra
eroded = erosion(orig_phantom, selem)

# Menampilkan hasil perbandingan
tampilkan(orig_phantom, eroded, 'erosion')

# Menampilkan nilai/bentuk elemen struktur
selem
```

```python
# Mendefinisikan elemen struktur berbentuk cakram (disk) dengan radius 19
strel_op = disk(19)

# Menerapkan operasi Closing pada citra
opened = closing(orig_phantom, strel_op)

# Menampilkan perbandingan citra asli dengan hasil closing
tampilkan(orig_phantom, opened, 'closing')
```

```python
# Menduplikasi citra asli untuk dimodifikasi
phantom = orig_phantom.copy()

# Memberikan gangguan (noise) buatan pada citra
phantom[340:350, 200:210] = 255  # Kotak putih cerah
phantom[100:110, 200:210] = 0    # Kotak hitam gelap

# White Top-Hat Transform
strl = disk(3)
w_tophat = white_tophat(phantom, strl)
tampilkan(phantom, w_tophat, 'white tophat')
```

```python
from skimage.util import invert

# Memuat data citra kuda dan membalikkan warnanya (Invert)
# Hal ini dilakukan agar objek kuda menjadi putih di atas latar belakang hitam
kuda = invert(data.horse())

# Menerapkan algoritma Skeletonize untuk mendapatkan kerangka objek
tulang = skeletonize(kuda)

# Menampilkan perbandingan antara citra kuda asli dan hasil skeletonize
tampilkan(kuda, tulang, 'Skeletonize')
```
